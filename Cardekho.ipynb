# 5.Data Modelling

# ---> 5.1 Preparing Data

```
df1["Current_year"]=2020
df1["no_of_Year"]=df1["Current_year"]- df1["Year"]
df1.drop(columns=["Current_year","Year"],axis=1,inplace=True)
df1.drop(columns=["Car_Name"],axis=1,inplace=True)
```

```
#Preprocessing
df1=pd.get_dummies(df1,drop_first="first")
df1.corr()
```

```
X=df1.iloc[:,1:9]
Y=df1.iloc[:,0]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=True)
```

```
X_train.shape
```

```
X_test.shape
```

```
Y_train.shape
```

```
Y_test.shape
```

# ---> 5.2 Training the Algorithm

```
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(X_train,Y_train)
```

```
print('Model Score : ',linear_model.score(X_train,Y_train))
```

```
coeff_df = pd.DataFrame(linear_model.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

# ---> 5.3 Making Prediction

```
prediction=linear_model.predict(X_test)
```

```
df = pd.DataFrame({'Actual': Y_test, 'Predicted': prediction})
df.head()
```

```
sns.distplot(Y_test - prediction)
```

```
sns.scatterplot(Y_test, prediction)
```

# ---> 5.4 Evaluating Algorithm

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, prediction))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, prediction))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, prediction)))
```

```
print('Model Score : ',linear_model.score(X_train,Y_train))
```
